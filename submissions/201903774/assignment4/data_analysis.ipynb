```python
# 4th Assignment: Data Collection and Analysis - 201903774 언어인지과학과 한형준
# 해당 데이터 분석 코드의 모든 코드 셀은 ChatGPT 5.1에 요청하여 생성되었음.

# ============================================================
# 0. 환경 설정 (Colab 전용)
# ============================================================

# 0-1) 라이브러리 설치 (한 번만 실행)
!pip install -q datasets pandas matplotlib seaborn

# 0-2) 나눔고딕 폰트 설치
!apt-get update -qq
!apt-get install -y fonts-nanum

import os
import math
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

from datasets import load_dataset

# 0-3) 나눔고딕 폰트 경로
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
print("폰트 파일 존재 여부:", os.path.exists(font_path))

if os.path.exists(font_path):
    # 폰트 매니저에 추가
    fm.fontManager.addfont(font_path)
    font_prop = fm.FontProperties(fname=font_path)
    nanum_name = font_prop.get_name()
    print("matplotlib에서 인식한 폰트 이름:", nanum_name)

    # 💡 전역 기본 폰트를 나눔고딕으로 강제 설정
    mpl.rcParams["font.family"] = nanum_name
    mpl.rcParams["font.sans-serif"] = [nanum_name]
    plt.rcParams["font.family"] = nanum_name
    plt.rcParams["font.sans-serif"] = [nanum_name]
    mpl.rcParams["axes.unicode_minus"] = False  # 마이너스 기호 깨짐 방지
else:
    print("⚠ NanumGothic.ttf를 찾지 못했습니다. DejaVu Sans로 fallback 됩니다.")

from matplotlib import font_manager as fm

kfont = fm.FontProperties(fname="/usr/share/fonts/truetype/nanum/NanumGothic.ttf")

# 과제 제출 구조에 맞춰 data 디렉토리 생성
os.makedirs("data", exist_ok=True)

# ============================================================
# 1. 데이터 수집: 한국어 & 영어 요약 데이터셋 로드
#    - 한국어: daekeun-ml/naver-news-summarization-ko (HF)
#    - 영어:   gopalkalpande/bbc-news-summary (HF)
# ============================================================

print("=== 1-1. 한국어 요약 데이터셋 로드 ===")
ko_ds = load_dataset("daekeun-ml/naver-news-summarization-ko")  # default split: train/validation/test
ko_train = ko_ds["train"].to_pandas()

print("한국어 train shape:", ko_train.shape)
print(ko_train.head(3))

# 주요 컬럼: date, category, press, title, document, link, summary 등
# document: 뉴스 본문, summary: 사람이 쓴 요약  :contentReference[oaicite:4]{index=4}

print("\n=== 1-2. 영어 요약 데이터셋 로드 ===")
en_ds = load_dataset("gopalkalpande/bbc-news-summary")  # 단일 split: train
en_train = en_ds["train"].to_pandas()

print("영어 train shape:", en_train.shape)
print(en_train.head(3))

# 주요 컬럼: File_path, Articles(본문), Summaries(요약)  :contentReference[oaicite:5]{index=5}

# ============================================================
# 2. 원본 데이터 CSV로 저장 (과제 제출용 data/ 폴더)
# ============================================================
ko_csv_path = "data/naver_news_summarization_ko.csv"
en_csv_path = "data/bbc_news_summary_en.csv"

ko_train.to_csv(ko_csv_path, index=False)
en_train.to_csv(en_csv_path, index=False)

print(f"\n[저장 완료] 한국어 데이터 -> {ko_csv_path}")
print(f"[저장 완료] 영어 데이터   -> {en_csv_path}")

# ============================================================
# 3. 데이터 기본 정보 확인 (info, describe, 결측치)
#    - Assignment 4 README에서 요구하는 최소 항목
# ============================================================

print("\n=== 3-1. 한국어 데이터 기본 정보 ===")
print(ko_train.info())
print("\n[한국어] 결측치 개수:")
print(ko_train.isna().sum())

print("\n=== 3-2. 영어 데이터 기본 정보 ===")
print(en_train.info())
print("\n[영어] 결측치 개수:")
print(en_train.isna().sum())

# 숫자형 칼럼에 대한 기본 통계(describe)는 참고용
print("\n=== 3-3. describe() ===")
print("[한국어] describe():")
print(ko_train.describe(include="all"))
print("\n[영어] describe():")
print(en_train.describe(include="all"))

# ============================================================
# 4. 텍스트 길이 특징 추가 (문자 수 기준)
#    - 본문 길이, 요약 길이, 압축 비율 등
# ============================================================

# 안전하게 get() + fillna("") 사용 (혹시라도 컬럼 누락 대비)
def safe_len(s):
    if isinstance(s, str):
        return len(s)
    return 0

# 한국어: document -> 본문, summary -> 요약
ko_df = ko_train.copy()
ko_df["doc_len"] = ko_df["document"].apply(safe_len)
ko_df["sum_len"] = ko_df["summary"].apply(safe_len)
ko_df["compression_ratio"] = ko_df["sum_len"] / ko_df["doc_len"].replace(0, np.nan)

# 영어: Articles -> 본문, Summaries -> 요약
en_df = en_train.copy()
en_df["doc_len"] = en_df["Articles"].apply(safe_len)
en_df["sum_len"] = en_df["Summaries"].apply(safe_len)
en_df["compression_ratio"] = en_df["sum_len"] / en_df["doc_len"].replace(0, np.nan)

print("\n=== 4-1. 한국어 텍스트 길이 예시 ===")
print(ko_df[["title", "doc_len", "sum_len", "compression_ratio"]].head(5))

print("\n=== 4-2. 영어 텍스트 길이 예시 ===")
print(en_df[["File_path", "doc_len", "sum_len", "compression_ratio"]].head(5))

print("\n=== 4-3. 레이블링 데이터셋 생성 (text, label 컬럼) + CSV 저장 ===")
#    - 한국어: text = document, label = category
#    - 영어:   text = Articles, label = File_path

# 한국어 레이블링
ko_labeled = pd.DataFrame({
    "text": ko_df["document"],
    "label": ko_df["category"]  # economy / IT과학 등
})

# 영어 레이블링
en_labeled = pd.DataFrame({
    "text": en_df["Articles"],
    "label": en_df["File_path"]  # politics / sport / business / tech / entertainment 등
})

# 저장
ko_labeled_path = "data/korean_text_label.csv"
en_labeled_path = "data/english_text_label.csv"

ko_labeled.to_csv(ko_labeled_path, index=False)
en_labeled.to_csv(en_labeled_path, index=False)

print(f"[저장 완료] 한국어 레이블링 데이터 -> {ko_labeled_path}")
print(f"[저장 완료] 영어 레이블링 데이터   -> {en_labeled_path}")

# 레이블 분포 확인
print("\n[한국어] label 분포 (category 기준):")
print(ko_labeled["label"].value_counts())

print("\n[영어] label 분포 (File_path 기준):")
print(en_labeled["label"].value_counts())

# ============================================================
# 5. 단어 빈도 분석을 위한 간단 토크나이저 & 유틸 함수
# ============================================================

import re
from collections import Counter

def tokenize_ko(text: str):
    """한국어: 한글/영문/숫자 + 공백만 남기고 나머지는 제거 후 공백 기준 토크나이즈"""
    if not isinstance(text, str):
        return []
    # 한글/영문/숫자/공백 외 제거
    text = re.sub(r"[^0-9A-Za-z가-힣\s]", " ", text)
    # 너무 짧은 토큰(1글자)은 제외
    tokens = [t for t in text.split() if len(t) > 1]
    return tokens

def tokenize_en(text: str):
    """영어: 소문자 변환 후 알파벳만 사용"""
    if not isinstance(text, str):
        return []
    text = text.lower()
    tokens = re.findall(r"[a-z]+", text)
    tokens = [t for t in tokens if len(t) > 1]
    return tokens

def top_n_words(texts, tokenizer, n=30, sample_size=None, random_state=0):
    """
    여러 문장 리스트에서 가장 많이 등장한 단어 n개 추출.
    sample_size가 주어지면 일부만 샘플링해서 속도 절약.
    """
    if sample_size is not None and len(texts) > sample_size:
        texts = texts.sample(sample_size, random_state=random_state)

    counter = Counter()
    for t in texts:
        counter.update(tokenizer(str(t)))
    return counter.most_common(n)

def top_n_words_by_label(df, tokenizer, n=20, sample_size=None, random_state=0):
    """
    label별로 가장 많이 등장한 단어 n개 추출.
    df는 text/label 컬럼을 가진 DataFrame.
    """
    results = {}
    labels = df["label"].unique()
    for lbl in labels:
        sub = df[df["label"] == lbl]["text"]
        if sample_size is not None and len(sub) > sample_size:
            sub = sub.sample(sample_size, random_state=random_state)
        counter = Counter()
        for t in sub:
            counter.update(tokenizer(str(t)))
        results[lbl] = counter.most_common(n)
    return results

# ============================================================
# 5-1. 전체 코퍼스에서 자주 등장하는 단어 (Top-N)
# ============================================================

# 연산량을 줄이기 위해 한국어는 최대 10,000개 문서, 영어는 최대 2,000개 문서만 샘플링
print("=== 한국어 전체 코퍼스 상위 단어 ===")
ko_top_words = top_n_words(
    ko_labeled["text"], tokenizer=tokenize_ko, n=30, sample_size=10000, random_state=42
)
for w, c in ko_top_words:
    print(f"{w}\t{c}")

print("\n=== 영어 전체 코퍼스 상위 단어 ===")
en_top_words = top_n_words(
    en_labeled["text"], tokenizer=tokenize_en, n=30, sample_size=2000, random_state=42
)
for w, c in en_top_words:
    print(f"{w}\t{c}")

# ============================================================
# 5-2. label(카테고리) 별로 자주 등장하는 단어
#    - 한국어: category 기준 (economy / IT과학 등)
#    - 영어:   File_path 기준 (politics / sport / business / tech / entertainment)
# ============================================================

print("=== 한국어: label(카테고리)별 상위 단어 ===")
ko_label_top = top_n_words_by_label(
    ko_labeled, tokenizer=tokenize_ko, n=20, sample_size=5000, random_state=42
)

for lbl, words in ko_label_top.items():
    print(f"\n[라벨: {lbl}] 상위 단어 20개")
    for w, c in words:
        print(f"  {w}\t{c}")

print("\n\n=== 영어: label(카테고리)별 상위 단어 ===")
en_label_top = top_n_words_by_label(
    en_labeled, tokenizer=tokenize_en, n=20, sample_size=1500, random_state=42
)

for lbl, words in en_label_top.items():
    print(f"\n[라벨: {lbl}] 상위 단어 20개")
    for w, c in words:
        print(f"  {w}\t{c}")


# ============================================================
# 6. 분포 시각화 (EDA 핵심)
#    - 본문 길이 히스토그램
#    - 요약 길이 히스토그램
#    - 압축 비율 분포
# ============================================================

sns.set_style("whitegrid")

# ---- 한국어 본문 길이 분포 ----
plt.figure()
ko_df["doc_len"].hist(bins=50)
plt.title("한국어 뉴스 본문 길이 분포 (문자 수)", fontproperties=kfont)
plt.xlabel("문자 수", fontproperties=kfont)
plt.ylabel("문서 개수", fontproperties=kfont)
plt.show()

# ---- 한국어 요약 길이 분포 ----
plt.figure()
ko_df["sum_len"].hist(bins=50)
plt.title("한국어 뉴스 요약 길이 분포 (문자 수)", fontproperties=kfont)
plt.xlabel("문자 수", fontproperties=kfont)
plt.ylabel("요약 개수", fontproperties=kfont)
plt.show()

# ---- 한국어 압축 비율 분포 ----
plt.figure()
ko_df["compression_ratio"].dropna().hist(bins=50)
plt.title("한국어 요약 압축 비율 분포 (요약 길이 / 본문 길이)", fontproperties=kfont)
plt.xlabel("압축 비율", fontproperties=kfont)
plt.ylabel("문서 개수", fontproperties=kfont)
plt.show()

# ---- 영어 본문 길이 분포 ----
plt.figure()
en_df["doc_len"].hist(bins=50)
plt.title("영어 뉴스 본문 길이 분포 (문자 수)", fontproperties=kfont)
plt.xlabel("문자 수", fontproperties=kfont)
plt.ylabel("문서 개수", fontproperties=kfont)
plt.show()

# ---- 영어 요약 길이 분포 ----
plt.figure()
en_df["sum_len"].hist(bins=50)
plt.title("영어 뉴스 요약 길이 분포 (문자 수)", fontproperties=kfont)
plt.xlabel("문자 수", fontproperties=kfont)
plt.ylabel("요약 개수", fontproperties=kfont)
plt.show()

# ---- 영어 압축 비율 분포 ----
plt.figure()
en_df["compression_ratio"].dropna().hist(bins=50)
plt.title("영어 요약 압축 비율 분포 (요약 길이 / 본문 길이)", fontproperties=kfont)
plt.xlabel("압축 비율", fontproperties=kfont)
plt.ylabel("문서 개수", fontproperties=kfont)
plt.show()

# ============================================================
# 7. 본문 길이 vs 요약 길이 관계 (산점도)
#    - 두 언어 모두에서 "요약이 어느 정도로 압축되어 있는지" 시각적으로 확인
# ============================================================

# 한국어: 극단적인 outlier 제거 (상위 1% 컷)
ko_q99_doc = ko_df["doc_len"].quantile(0.99)
ko_q99_sum = ko_df["sum_len"].quantile(0.99)
ko_scatter = ko_df[(ko_df["doc_len"] <= ko_q99_doc) & (ko_df["sum_len"] <= ko_q99_sum)]

plt.figure()
plt.scatter(ko_scatter["doc_len"], ko_scatter["sum_len"], alpha=0.3)
plt.title("한국어: 본문 길이 vs 요약 길이 (상위 1% 제거)", fontproperties=kfont)
plt.xlabel("본문 길이 (문자 수)", fontproperties=kfont)
plt.ylabel("요약 길이 (문자 수)", fontproperties=kfont)
plt.show()

# 영어: 마찬가지로 상위 1% 컷
en_q99_doc = en_df["doc_len"].quantile(0.99)
en_q99_sum = en_df["sum_len"].quantile(0.99)
en_scatter = en_df[(en_df["doc_len"] <= en_q99_doc) & (en_df["sum_len"] <= en_q99_sum)]

plt.figure()
plt.scatter(en_scatter["doc_len"], en_scatter["sum_len"], alpha=0.3)
plt.title("영어: 본문 길이 vs 요약 길이 (상위 1% 제거)", fontproperties=kfont)
plt.xlabel("본문 길이 (문자 수)", fontproperties=kfont)
plt.ylabel("요약 길이 (문자 수)", fontproperties=kfont)
plt.show()

# ============================================================
# 8. 샘플 확인: 실제 본문 + 요약 몇 개 출력
#    - BonCahier AI 프로젝트 취지(긴 강의 슬라이드 → 짧은 요약)에 맞는지 직관적으로 확인
# ============================================================

def print_ko_example(idx=0):
    row = ko_df.iloc[idx]
    print("="*80)
    print(f"[한국어 예시 #{idx}]")
    print(f"제목: {row['title']}")
    print(f"카테고리: {row['category']}, 언론사: {row['press']}")
    print(f"- 본문(일부, 앞 400자):\n{row['document'][:400]}...\n")
    print(f"- 요약:\n{row['summary']}\n")
    print(f"(본문 길이: {row['doc_len']}, 요약 길이: {row['sum_len']}, 압축비: {row['compression_ratio']:.3f})")

def print_en_example(idx=0):
    row = en_df.iloc[idx]
    print("="*80)
    print(f"[영어 예시 #{idx}]")
    print(f"파일 경로: {row['File_path']}")
    print(f"- 본문(일부, 앞 400자):\n{row['Articles'][:400]}...\n")
    print(f"- 요약:\n{row['Summaries']}\n")
    print(f"(본문 길이: {row['doc_len']}, 요약 길이: {row['sum_len']}, 압축비: {row['compression_ratio']:.3f})")

print_ko_example(0)
print_en_example(0)

# ============================================================
# 9. 간단한 텍스트 요약 (Markdown 형태로 정리할 때 써먹기 좋은 내용)
#     - 이 셀의 출력 내용을 README.md에 옮겨 요약 문단으로 활용 가능
# ============================================================

print("\n\n================= 요약 정리용 메모 =================")
print(f"- 한국어 데이터 개수 (train): {len(ko_df)}")
print(f"  * 본문 평균 길이: {ko_df['doc_len'].mean():.1f} 문자")
print(f"  * 요약 평균 길이: {ko_df['sum_len'].mean():.1f} 문자")
print(f"  * 평균 압축 비율: {ko_df['compression_ratio'].mean():.3f}")

print("\n- 영어 데이터 개수 (train):", len(en_df))
print(f"  * 본문 평균 길이: {en_df['doc_len'].mean():.1f} 문자")
print(f"  * 요약 평균 길이: {en_df['sum_len'].mean():.1f} 문자")
print(f"  * 평균 압축 비율: {en_df['compression_ratio'].mean():.3f}")
```
